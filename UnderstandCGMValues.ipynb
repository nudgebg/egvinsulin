```python
import pandas as pd
```

```python
#load all cgm data
cgm_data_IOBP2 = pd.read_csv('data/raw/IOBP2 RCT Public Dataset/Data Tables/IOBP2DeviceiLet.txt', sep='|')
cgm_data_DCLP3 = pd.read_csv('data/raw/DCLP3 Public Dataset - Release 3 - 2022-08-04/Data Files/Pump_CGMGlucoseValue.txt', sep='|')
cgm_data_DCLP5 = pd.read_csv('data/raw/DCLP5_Dataset_2022-01-20-5e0f3b16-c890-4ace-9e3b-531f3687cf53/DCLP5TandemCGMDATAGXB_b.txt', sep='|')
cgm_data_PEDAP = pd.read_csv('data/raw/PEDAP Public Dataset - Release 3 - 2024-09-25/Data Files/PEDAPTandemCGMDATAGXB.txt', sep='|')
```

```python
#print all unique cgm values outside of the normal range of 40 - 400 for each study
print('IOBP2')
print(cgm_data_IOBP2[(cgm_data_IOBP2['CGMVal'] < 40) | (cgm_data_IOBP2['CGMVal'] > 400)]['CGMVal'].unique())
print('DCLP3')
print(cgm_data_DCLP3[(cgm_data_DCLP3['CGMValue'] < 40) | (cgm_data_DCLP3['CGMValue'] > 400)]['CGMValue'].unique())
print('DCLP5')
print(cgm_data_DCLP5[(cgm_data_DCLP5['CGMValue'] < 40) | (cgm_data_DCLP5['CGMValue'] > 400)]['CGMValue'].unique())
print('PEDAP')
print(cgm_data_PEDAP[(cgm_data_PEDAP['CGMValue'] < 40) | (cgm_data_PEDAP['CGMValue'] > 400)]['CGMValue'].unique())
```
